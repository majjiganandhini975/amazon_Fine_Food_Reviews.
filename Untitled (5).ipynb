```python
df["ProfileName"].fillna("Unknown", inplace=True)
df["Summary"].fillna("No Summary", inplace=True)
```

```python
#checking duplicates

df.duplicated().sum()
```

```python
#removing duplicates.

df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
#HANDLE WRONG DATA

#age cannot be negative

df[df["HelpfulnessDenominator"] < 0]
```

```python
#removing invalid rows

df = df[df["Score"] > 0]
df
```
